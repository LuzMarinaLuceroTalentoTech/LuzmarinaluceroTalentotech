```python
# Conjunto A
conjunto_A = {"manzana", "plátano", "cereza", "naranja"}

# Conjunto B
conjunto_B = {"mango", "naranja", "manzana", "kiwi"}

# Realizar la intersección de ambos conjuntos
interseccion = conjunto_A.intersection(conjunto_B)

# Mostrar el resultado de la intersección
print("Conjunto A:", conjunto_A)
print("Conjunto B:", conjunto_B)
print("Intersección de A y B:", interseccion)

```

```python
from tabulate import tabulate

# Lista de productos
productos = [
    {"Producto": "Arroz", "Cantidad": 2, "Precio Unitario": 1.5},
    {"Producto": "Azúcar", "Cantidad": 1, "Precio Unitario": 2.0},
    {"Producto": "Café", "Cantidad": 3, "Precio Unitario": 3.5},
    {"Producto": "Chocolate", "Cantidad": 2, "Precio Unitario": 4.0},
]

# Calcular el precio total para cada producto
for producto in productos:
    producto["Precio Total"] = producto["Cantidad"] * producto["Precio Unitario"]

# Calcular el total de la factura
total_factura = sum(producto["Precio Total"] for producto in productos)

# Crear la tabla
tabla = tabulate(productos, headers="keys", tablefmt="grid")

# Imprimir la tabla y el total
print(tabla)
print(f"\nTotal de la factura: ${total_factura:.2f}")
```

```python
import random

def adivina_el_numero():
    print("¡Bienvenido al juego de Adivina el Número!")
    print("Estoy pensando en un número entre 1 y 100.")

    # Elige un número aleatorio entre 1 y 100
    numero_secreto = random.randint(1, 100)
    intentos = 0

    while True:
        # Pide al jugador que adivine el número
        try:
            adivinanza = int(input("Adivina el número: "))
        except ValueError:
            print("Por favor, ingresa un número válido.")
            continue

        intentos += 1

        # Comprueba si la adivinanza es correcta
        if adivinanza < numero_secreto:
            print("Muy bajo. Intenta nuevamente.")
        elif adivinanza > numero_secreto:
            print("Muy alto. Intenta nuevamente.")
        else:
            print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
            break

# Ejecutar el juego
adivina_el_numero()
```

```python
import pandas as pd

# Crear una lista de diccionarios con los elementos de la canasta familiar, sus precios unitarios, cantidad y unidad de medida
canasta_familiar = [
    {'Producto': 'Arroz', 'Precio Unitario': 3500, 'Cantidad': 2, 'Unidad': 'kg'},
    {'Producto': 'Frijoles', 'Precio Unitario': 4800, 'Cantidad': 1, 'Unidad': 'kg'},
    {'Producto': 'Azúcar', 'Precio Unitario': 2200, 'Cantidad': 3, 'Unidad': 'kg'},
    {'Producto': 'Aceite', 'Precio Unitario': 8900, 'Cantidad': 1, 'Unidad': 'L'},
    {'Producto': 'Leche', 'Precio Unitario': 3000, 'Cantidad': 4, 'Unidad': 'L'},
    {'Producto': 'Huevos', 'Precio Unitario': 700, 'Cantidad': 30, 'Unidad': 'unidades'},
    {'Producto': 'Pan', 'Precio Unitario': 1200, 'Cantidad': 6, 'Unidad': 'kg'},
    {'Producto': 'Queso', 'Precio Unitario': 8000, 'Cantidad': 1, 'Unidad': 'kg'},
    {'Producto': 'Carne', 'Precio Unitario': 15000, 'Cantidad': 2, 'Unidad': 'kg'},
    {'Producto': 'Jabón', 'Precio Unitario': 2500, 'Cantidad': 3, 'Unidad': 'unidades'}
]

# Crear un DataFrame a partir de la lista de diccionarios
df_canasta = pd.DataFrame(canasta_familiar)

# Añadir una columna para numerar las filas de 1 a 10
df_canasta.index = range(1, len(df_canasta) + 1)

# Calcular el valor total por producto
df_canasta['Total'] = df_canasta['Precio Unitario'] * df_canasta['Cantidad']

# Mostrar el DataFrame
print(df_canasta)

# Calcular el valor total de todos los productos en la canasta familiar
valor_total = df_canasta['Total'].sum()

# Mostrar el valor total
print("\nValor total de la canasta familiar: $", valor_total)


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos inventado
data = {
    'Producto': ['Arroz', 'Frijoles', 'Azúcar', 'Aceite', 'Leche', 'Huevos', 'Pan', 'Queso', 'Carne', 'Jabón'],
    'Ventas': [150, 90, 200, 120, 300, 250, 80, 60, 110, 170],
    'Precio Unitario': [3500, 4800, 2200, 8900, 3000, 700, 1200, 8000, 15000, 2500],
    'Categoría': ['Grano', 'Grano', 'Dulce', 'Aceite', 'Lácteo', 'Proteína', 'Panadería', 'Lácteo', 'Proteína', 'Limpieza']
}

df = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame
print("Vista inicial de los datos:")
print(df.head())

# Descripción estadística básica
print("\nDescripción estadística básica:")
print(df.describe())

# Visualización de la distribución de ventas
plt.figure(figsize=(8, 6))
sns.histplot(df['Ventas'], kde=True)
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de barras para las ventas por categoría
plt.figure(figsize=(8, 6))
sns.barplot(x='Categoría', y='Ventas', data=df, ci=None)
plt.title('Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.show()

# Análisis de correlación entre precio unitario y ventas
print("\nCorrelación entre Precio Unitario y Ventas:")
print(df[['Precio Unitario', 'Ventas']].corr())

# Gráfico de dispersión para la relación entre precio unitario y ventas
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Precio Unitario', y='Ventas', data=df)
plt.title('Relación entre Precio Unitario y Ventas')

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo sobre alimentación, energía y desempeño estudiantil
data = {
    'Estudiante': ['Estudiante 1', 'Estudiante 2', 'Estudiante 3', 'Estudiante 4', 'Estudiante 5', 'Estudiante 6', 'Estudiante 7', 'Estudiante 8', 'Estudiante 9', 'Estudiante 10'],
    'Calidad de Alimentación (1-5)': [5, 3, 4, 2, 5, 3, 4, 1, 5, 2],
    'Nivel de Energía (1-5)': [5, 3, 4, 2, 5, 3, 4, 1, 5, 2],
    'Promedio Académico (1-100)': [90, 75, 85, 65, 92, 78, 88, 55, 94, 70]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Excluir la columna 'Estudiante' al calcular la matriz de correlación
df_correlation_matrix = df.drop(columns=['Estudiante']).corr()

# Mostrar la matriz de correlación como tabla utilizando print o display
print("Matriz de Correlación:")
print(df_correlation_matrix)


# Visualizar la matriz de correlación como un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(df_correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación entre Alimentación, Energía y Desempeño Académico')
plt.show()
df_correlation_matrix
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

# Valores reales (True Labels)
y_true = ['Spam', 'No Spam', 'No Spam', 'No Spam', 'Spam', 'No Spam', 'Spam', 'No Spam', 'Spam', 'Spam']

# Predicciones del modelo (Predicted Labels)
y_pred = ['Spam', 'No Spam', 'Spam', 'No Spam', 'Spam', 'Spam', 'No Spam', 'No Spam', 'Spam', 'No Spam']

# Crear la matriz de confusión
cm = confusion_matrix(y_true, y_pred, labels=['Spam', 'No Spam'])

# Convertir la matriz a un DataFrame para una presentación más clara
df_cm = pd.DataFrame(cm, index=['Actual Spam', 'Actual No Spam'], columns=['Predicted Spam', 'Predicted No Spam'])

# Mostrar la matriz de confusión
print(df_cm)
```
