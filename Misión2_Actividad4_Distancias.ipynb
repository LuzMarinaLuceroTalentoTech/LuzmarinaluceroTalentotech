En el caso de puntos en n dimensiones, la f贸rmula se extiende sumando la diferencia al cuadrado para cada coordenada.

```python
import math

def euclidean_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Both points must have the same number of dimensions")

    distance = math.sqrt(sum((p1 - p2) ** 2 for p1, p2 in zip(point1, point2)))
    return distance

# Ejemplo de uso
pointA = (1, 2)
pointB = (4, 6)
dist = euclidean_distance(pointA, pointB)
print(f"La distancia euclidiana es: {dist}")
```

```python
import math
import matplotlib.pyplot as plt

def manhattan_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Both points must have the same number of dimensions")

    distance = sum(abs(p1 - p2) for p1, p2 in zip(point1, point2))
    return distance

# Puntos de ejemplo
pointA = (1, 2)
pointB = (4, 6)

# Calcular la distancia de Manhattan
manhattan_dist = manhattan_distance(pointA, pointB)
print(f"La distancia de Manhattan es: {manhattan_dist}")

# Graficar los puntos y la l铆nea de distancia
plt.figure(figsize=(6, 6))

# Dibujar los puntos
plt.scatter(*pointA, color='blue', label='Punto A (1,2)', s=100)
plt.scatter(*pointB, color='red', label='Punto B (4,6)', s=100)

# Dibujar la l铆nea que conecta los puntos
# Calculamos los puntos intermedios para crear el recorrido
midpoint1 = (pointA[0], pointB[1])  # Movimiento vertical
midpoint2 = (pointB[0], pointB[1])  # Movimiento horizontal

plt.plot([pointA[0], midpoint1[0]], [pointA[1], midpoint1[1]], color='green', linestyle='--')
plt.plot([midpoint1[0], pointB[0]], [midpoint1[1], pointB[1]], color='green', linestyle='--')

# A帽adir etiquetas y leyenda
plt.text(pointA[0], pointA[1], f'  A{pointA}', fontsize=12, color='blue')
plt.text(pointB[0], pointB[1], f'  B{pointB}', fontsize=12, color='red')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Distancia de Manhattan entre dos puntos')
plt.legend()

# Mostrar la gr谩fica
plt.grid(True)
plt.show()
```

La distancia de Hamming es una m茅trica que se utiliza para comparar dos cadenas de igual longitud. Cuenta el n煤mero de posiciones en las que los valores de las dos cadenas son diferentes. Es com煤nmente usada en teor铆a de la informaci贸n y an谩lisis de secuencias de bits.

```python
def hamming_distance(seq1, seq2):
    if len(seq1) != len(seq2):
        raise ValueError("Las secuencias deben tener la misma longitud")

    # Contar el n煤mero de posiciones donde los valores son diferentes
    distance = sum(el1 != el2 for el1, el2 in zip(seq1, seq2))
    return distance

# Ejemplo de uso
seqA = "1011001"
seqB = "1001001"

dist_hamming = hamming_distance(seqA, seqB)
print(f"La distancia de Hamming entre {seqA} y {seqB} es: {dist_hamming}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def hamming_distance(seq1, seq2):
    if len(seq1) != len(seq2):
        raise ValueError("Las secuencias deben tener la misma longitud")

    # Encontrar las posiciones donde los elementos son diferentes
    differences = [i for i, (el1, el2) in enumerate(zip(seq1, seq2)) if el1 != el2]

    distance = len(differences)
    return distance, differences

# Ejemplo de uso
seqA = "1011001"
seqB = "1001001"

# Calcular la distancia de Hamming y las posiciones donde son diferentes
dist_hamming, diff_positions = hamming_distance(seqA, seqB)
print(f"La distancia de Hamming entre {seqA} y {seqB} es: {dist_hamming}")

# Graficar las diferencias
x = np.arange(len(seqA))  # Ejes X con la longitud de las secuencias
y_seqA = [int(bit) for bit in seqA]
y_seqB = [int(bit) for bit in seqB]

plt.figure(figsize=(8, 4))

# Graficar secuencia A y B
plt.plot(x, y_seqA, marker='o', label='Secuencia A', color='blue')
plt.plot(x, y_seqB, marker='o', label='Secuencia B', color='red')

# Resaltar las diferencias
for pos in diff_positions:
    plt.axvline(x=pos, color='green', linestyle='--', label=f'Diferencia en posici贸n {pos+1}')

# A帽adir detalles a la gr谩fica
plt.xticks(x)
plt.yticks([0, 1], ['0', '1'])
plt.xlabel('Posici贸n en la secuencia')
plt.ylabel('Valor binario')
plt.title(f'Distancia de Hamming: {dist_hamming} diferencias')
plt.legend(loc='best')
plt.grid(True)

# Mostrar la gr谩fica
plt.show()
```

```python
import matplotlib.pyplot as plt

def chebyshev_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Ambos puntos deben tener el mismo n煤mero de dimensiones")

    # Calculamos la distancia de Chebyshev
    distance = max(abs(p1 - p2) for p1, p2 in zip(point1, point2))
    return distance

# Puntos de ejemplo
pointA = (1, 2)
pointB = (4, 6)

# Calcular la distancia de Chebyshev
dist_chebyshev = chebyshev_distance(pointA, pointB)
print(f"La distancia de Chebyshev entre {pointA} y {pointB} es: {dist_chebyshev}")

# Graficar los puntos y la distancia de Chebyshev
plt.figure(figsize=(6, 6))

# Dibujar los puntos
plt.scatter(*pointA, color='blue', label=f'Punto A {pointA}', s=100)
plt.scatter(*pointB, color='red', label=f'Punto B {pointB}', s=100)

# Dibujar el recorrido "m谩ximo"
plt.plot([pointA[0], pointB[0]], [pointA[1], pointA[1]], color='green', linestyle='--')
plt.plot([pointB[0], pointB[0]], [pointA[1], pointB[1]], color='green', linestyle='--')

# A帽adir etiquetas y leyenda
plt.text(pointA[0], pointA[1], f'  A{pointA}', fontsize=12, color='blue')
plt.text(pointB[0], pointB[1], f'  B{pointB}', fontsize=12, color='red')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title(f'Distancia de Chebyshev: {dist_chebyshev}')
plt.legend()

# Mostrar la gr谩fica
plt.grid(True)
plt.show()
```

```python
import numpy as np

def canberra_distance(point1, point2):
    if len(point1) != len(point2):
        raise ValueError("Ambos puntos deben tener el mismo n煤mero de dimensiones")

    distance = sum(abs(p1 - p2) / (abs(p1) + abs(p2)) if (p1 != 0 or p2 != 0) else 0
                   for p1, p2 in zip(point1, point2))
    return distance

# Puntos de ejemplo
pointA = [1, 2, 3]
pointB = [2, 4, 6]

# Calcular la distancia de Canberra
dist_canberra = canberra_distance(pointA, pointB)
print(f"La distancia de Canberra entre {pointA} y {pointB} es: {dist_canberra}")
```

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    # Producto punto de los dos vectores
    dot_product = np.dot(vec1, vec2)

    # Normas de los vectores
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)

    # Calcular la similitud coseno
    cosine_sim = dot_product / (norm_vec1 * norm_vec2)
    return cosine_sim

def cosine_distance(vec1, vec2):
    # La distancia coseno es 1 menos la similitud coseno
    return 1 - cosine_similarity(vec1, vec2)

# Vectores de ejemplo
vecA = np.array([1, 2, 3])
vecB = np.array([4, 5, 6])

# Calcular la distancia coseno
dist_cosine = cosine_distance(vecA, vecB)
print(f"La distancia coseno entre {vecA} y {vecB} es: {dist_cosine:.4f}")
```

```python
def jaccard_similarity(set1, set2):
    # Calcular la intersecci贸n y la uni贸n de los conjuntos
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))

    # Calcular la similitud de Jaccard
    similarity = intersection / union
    return similarity

def jaccard_distance(set1, set2):
    # La distancia de Jaccard es 1 menos la similitud de Jaccard
    return 1 - jaccard_similarity(set1, set2)

# Conjuntos de ejemplo
setA = {1, 2, 3, 4, 5}
setB = {4, 5, 6, 7, 8}

# Calcular la distancia de Jaccard
dist_jaccard = jaccard_distance(setA, setB)
print(f"La distancia de Jaccard entre {setA} y {setB} es: {dist_jaccard:.4f}")
```
