````python
from tabulate import tabulate

# Lista de productos
productos = [
    {"Producto": "Arroz", "Cantidad": 2, "Precio Unitario": 1.5},
    {"Producto": "Azúcar", "Cantidad": 1, "Precio Unitario": 2.0},
    {"Producto": "Café", "Cantidad": 3, "Precio Unitario": 3.5},
    {"Producto": "Chocolate", "Cantidad": 2, "Precio Unitario": 4.0},
]

# Calcular el precio total para cada producto
for producto in productos:
    producto["Precio Total"] = producto["Cantidad"] * producto["Precio Unitario"]

# Calcular el total de la factura
total_factura = sum(producto["Precio Total"] for producto in productos)

# Crear la tabla
tabla = tabulate(productos, headers="keys", tablefmt="grid")

# Imprimir la tabla y el total
print(tabla)
print(f"\nTotal de la factura: ${total_factura:.2f}")
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un conjunto de datos inventado
data = {
    'Producto': ['Arroz', 'Frijoles', 'Azúcar', 'Aceite', 'Leche', 'Huevos', 'Pan', 'Queso', 'Carne', 'Jabón'],
    'Ventas': [150, 90, 200, 120, 300, 250, 80, 60, 110, 170],
    'Precio Unitario': [3500, 4800, 2200, 8900, 3000, 700, 1200, 8000, 15000, 2500],
    'Categoría': ['Grano', 'Grano', 'Dulce', 'Aceite', 'Lácteo', 'Proteína', 'Panadería', 'Lácteo', 'Proteína', 'Limpieza']
}

df = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame
print("Vista inicial de los datos:")
print(df.head())

# Descripción estadística básica
print("\nDescripción estadística básica:")
print(df.describe())

# Visualización de la distribución de ventas
plt.figure(figsize=(8, 6))
sns.histplot(df['Ventas'], kde=True)
plt.title('Distribución de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.show()

# Gráfico de barras para las ventas por categoría
plt.figure(figsize=(8, 6))
sns.barplot(x='Categoría', y='Ventas', data=df, ci=None)
plt.title('Ventas por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas')
plt.show()

# Análisis de correlación entre precio unitario y ventas
print("\nCorrelación entre Precio Unitario y Ventas:")
print(df[['Precio Unitario', 'Ventas']].corr())

# Gráfico de dispersión para la relación entre precio unitario y ventas
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Precio Unitario', y='Ventas', data=df)
plt.title('Relación entre Precio Unitario y Ventas')
plt.xlabel('Precio Unitario')
plt.ylabel('Ventas')
plt.show()
````

Explicación del Ejemplo:
Carga y vista inicial de datos: Se crea un pequeño DataFrame con productos, ventas, precios unitarios, y categorías.

Descripción estadística: df.describe() proporciona estadísticas básicas como la media, la desviación estándar, los valores mínimos y máximos, etc.

Visualización de la distribución: Usamos sns.histplot para visualizar cómo se distribuyen las ventas entre los productos.

Análisis por categoría: Se puede ver cómo se distribuyen las ventas según las categorías de productos con un gráfico de barras (sns.barplot).

Correlación y visualización de relaciones: Analizamos la relación entre el precio unitario y las ventas, mostrando la correlación y visualizándola con un gráfico de dispersión.

# A more complete example with synthetic data given by ChatGPT

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)  # for reproducibility

data = {
    'A': np.random.randn(100),
    'B': np.random.randn(100) * 2,
    'C': np.random.randn(100) + 1,
    'D': np.random.randn(100) - 1,
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
````
