```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Crear un DataFrame con los datos simulados
data = {
    'Tamaño': [50, 60, 70, 80, 90, 100, 110, 120, 130, 140],
    'Habitaciones': [1, 2, 2, 3, 3, 3, 4, 4, 5, 5],
    'Precio': [200, 230, 250, 270, 300, 330, 350, 370, 400, 420]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Definir las variables independientes (Tamaño y Habitaciones)
X = df[['Tamaño', 'Habitaciones']]

# Definir la variable dependiente (Precio)
Y = df['Precio']

# Añadir una constante a las variables independientes
X = sm.add_constant(X)

# Ajustar el modelo de regresión múltiple
modelo = sm.OLS(Y, X).fit()

# Mostrar el resumen del modelo
modelo.summary()

# Predecir los precios usando el modelo
predicciones = modelo.predict(X)

# Mostrar las predicciones
predicciones

# Graficar los valores reales vs. predichos
plt.figure(figsize=(8, 6))
plt.scatter(df['Tamaño'], Y, color='blue', label='Valores Reales')
plt.scatter(df['Tamaño'], predicciones, color='red', label='Valores Predichos')
plt.xlabel('Tamaño de la Vivienda (m²)')
plt.ylabel('Precio de la Vivienda')
plt.title('Regresión Múltiple: Precio vs. Tamaño y Habitaciones')
plt.legend()
plt.show()

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(df[['Tamaño', 'Habitaciones']], Y, test_size=0.2, random_state=0)

# Ajustar el modelo con scikit-learn
modelo_sklearn = LinearRegression()
modelo_sklearn.fit(X_train, Y_train)

# Hacer predicciones sobre los datos de prueba
Y_pred = modelo_sklearn.predict(X_test)

# Evaluar el modelo
error_cuadratico_medio = mean_squared_error(Y_test, Y_pred)
print(f"Error Cuadrático Medio: {error_cuadratico_medio}")



```

```python
# Importar librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Cargar el dataset de California Housing
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['MedHouseValue'] = california.target  # MedHouseValue es la variable dependiente (valor medio de la vivienda)

# Mostrar las primeras filas del dataset
df.head()
```

```python
# Descripción estadística básica del dataset
print(df.describe())

# Verificar si hay valores nulos
print(df.isnull().sum())

# Matriz de correlación
correlation_matrix = df.corr()

# Visualización de la matriz de correlación con un mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación de las Variables')
plt.show()
```

```python
# Definir las variables independientes (X) y dependiente (Y)
X = df.drop('MedHouseValue', axis=1)
Y = df['MedHouseValue']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X_train, Y_train)

# Hacer predicciones
Y_pred = modelo.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Error cuadrático medio (MSE): {mse}')
print(f'Coeficiente de determinación (R²): {r2}')

# Mostrar los coeficientes (pendientes)
coeficientes = pd.DataFrame(modelo.coef_, X.columns, columns=['Coeficiente'])
print(coeficientes)
```

```python
# Gráfica de valores reales vs predichos
plt.figure(figsize=(10, 6))
plt.scatter(Y_test, Y_pred, color='blue')
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linewidth=2)
plt.xlabel('Valores Reales')
plt.ylabel('Valores Predichos')
plt.title('Valores Reales vs Predichos')
plt.show()
```
