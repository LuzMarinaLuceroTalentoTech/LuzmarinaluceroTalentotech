```python
import pandas as pd

# URL de la API
url = "https://www.datos.gov.co/resource/sbwg-7ju4.json"

# Cargar los datos desde la URL
data = pd.read_json(url)

# Tomar una muestra aleatoria de 200 registros
data_sample = data.sample(n=200, random_state=42)

# Mostrar las primeras filas de la muestra para verificar
print(data_sample.head())
```

Análisis Estadístico.

```python
# Calcular estadísticas descriptivas
mean_value = data_sample['valorobservado'].mean()
median_value = data_sample['valorobservado'].median()
std_dev = data_sample['valorobservado'].std()
min_value = data_sample['valorobservado'].min()
max_value = data_sample['valorobservado'].max()

print("Análisis Estadístico de la Precipitación Observada:")
print(f"Media: {mean_value:.2f}")
print(f"Mediana: {median_value:.2f}")
print(f"Desviación Estándar: {std_dev:.2f}")
print(f"Mínimo: {min_value:.2f}")
print(f"Máximo: {max_value:.2f}")
```

Análisis Estadístico de la Precipitación Observada

Para comprender mejor el comportamiento de la precipitación en Colombia, se tomó una muestra aleatoria de 200 registros del conjunto de datos disponibles en el portal datos.gov.co . A partir de esta muestra, se realizó un análisis estadístico de la variable Precipitación Observada , obteniendo los siguientes resultados clave:

Media : 18,69 mm
Mediana : 18,29 mm
Desviación Estándar : 5,69 mm
Mínimo : 5,74 mm
Máximo : 33,00 mm

Estos valores indican una variabilidad moderada en las precipitaciones observadas en la muestra, con una media de 18.69 mm y una desviación estándar de 5.69 mm, lo cual sugiere que la mayoría de los valores se encuentran relativamente cerca del promedio. El valor mínimo de 5,74 mm muestra que hay períodos o regiones en la muestra con escasas precipitaciones, mientras que el máximo de 33,00 mm refleja episodios de lluvia más intensa.

La media y mediana similar (18.69 y 18.29 mm, respectivamente) sugiere que la distribución de los datos es aproximadamente simétrica y no presenta una gran cantidad de valores extremos en la muestra. Sin embargo, el rango de 5.74 mm a 33.00 mm revela que las precipitaciones pueden variar significativamente en diferentes zonas o periodos, lo cual es consistente con la variabilidad climática que se espera en Colombia.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Dividir en variables predictoras y objetivo
X = data_sample[['latitud', 'longitud']]
y = data_sample['valorobservado']

# División en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo de regresión lineal
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)

# Predicción y evaluación
y_pred_linear = model_linear.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear)

print("Error Cuadrático Medio (MSE) del Modelo de Regresión Lineal:", mse_linear)

# Modelo de Regresión Polinómica
from sklearn.preprocessing import PolynomialFeatures

# Generar términos polinómicos de grado 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# División en conjunto de entrenamiento y prueba
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.3, random_state=42)

# Modelo de regresión lineal con términos polinómicos
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)

# Predicción y evaluación
y_pred_poly = model_poly.predict(X_test_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)

print("Error Cuadrático Medio (MSE) del Modelo de Regresión Polinómica:", mse_poly)
```

Desempeño del Modelo de Regresión Lineal:

Se implementó un modelo de regresión lineal como un enfoque inicial para predecir la precipitación observada en función de las variables disponibles en el conjunto de datos. El modelo arrojó un Error Cuadrático Medio (MSE) de 27.36. Este valor de MSE indica una cantidad moderada de error en las predicciones, lo cual sugiere que el modelo de regresión lineal capta algunas de las relaciones presentes en los datos, pero tiene limitaciones al capturar la variabilidad de la precipitación, especialmente en valores extremos.

El uso de regresión lineal implica que la relación entre las variables predictoras y la precipitación observada es predominantemente lineal. Sin embargo, dado el error relativamente alto, es posible que la recepción esté influenciada por relaciones no lineales que no son bien capturadas por este modelo.

Desempeño del Modelo de Regresión Polinómica:

Para abordar posibles relaciones no lineales, se implementó un modelo de regresión polinómica . Este modelo tiene una mayor flexibilidad al agregar términos de mayor orden, permitiéndole capturar relaciones más complejas entre las variables predictoras y el resultado de precipitación. El modelo de regresión polinómica logró un MSE de 24.52, un valor ligeramente menor que el obtenido con la regresión lineal.
Esta mejora en el MSE, aunque marginal, sugiere que la incorporación de términos polinómicos permite al modelo ajustarse mejor a las fluctuaciones en los datos. Este resultado refuerza la idea de que existen patrones no lineales en la precipitación observada que no son capturados adecuadamente por la regresión lineal.

Comparación y Análisis de los Modelos

Aunque el modelo de regresión polinómica mostró un rendimiento ligeramente superior al modelo de regresión lineal , la reducción en el MSE no es sustancial. Esto indica que, si bien la relación entre las variables y la precipitación no es puramente lineal, la regresión polinómica no logra una mejora significativa, lo cual podría sugerir:
La posible presencia de factores adicionales o relaciones complejas que no son capturadas adecuadamente por los modelos lineales y polinómicos.
La necesidad de explorar modelos más avanzados, como interpolación espacial, que puedan capturar interacciones no lineales más complejas entre las variables.


```python
# Convertir variables categóricas a numéricas (si es necesario)
# En este ejemplo, consideramos "latitud" y "longitud" como características, y "valorobservado" como la variable objetivo

X = data[['latitud', 'longitud']]  # Características
y = data['valorobservado'] > data['valorobservado'].median()  # Etiqueta binaria (alta o baja)

# División de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Entrenar el Modelo de Clasificación: Usaremos un modelo básico de RandomForestClassifierpara realizar la clasificación.

```python
# Entrenar el modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)
```

Evaluación del Modelo: Calculamos métricas y visualizamos una matriz de confusión.

```python
# Reporte de clasificación y matriz de confusión
print(classification_report(y_test, y_pred))

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Modelo Interpolación Espacial
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Crear malla de puntos en el área de interés
grid_lat = np.linspace(data_sample['latitud'].min(), data_sample['latitud'].max(), 100)
grid_lon = np.linspace(data_sample['longitud'].min(), data_sample['longitud'].max(), 100)
grid_lat, grid_lon = np.meshgrid(grid_lat, grid_lon)

# Interpolación IDW
points = data_sample[['latitud', 'longitud']].values
values = data_sample['valorobservado'].values
grid_z = griddata(points, values, (grid_lat, grid_lon), method='cubic')

# Graficar resultados de interpolación
plt.figure(figsize=(10, 8))
plt.contourf(grid_lon, grid_lat, grid_z, cmap='viridis')
plt.colorbar(label='Precipitación Estimada')
plt.scatter(data_sample['longitud'], data_sample['latitud'], c='red', edgecolor='k', s=100, label='Puntos de Observación')
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.title("Interpolación Espacial de Precipitación (IDW)")
plt.show()
```

Rango de Precipitación:

La barra de color en la derecha indica que los valores de precipitación interpolada varían aproximadamente entre -15 y 105 mm.
Las áreas de color amarillo y verde claro representan las zonas con los niveles de precipitación estimados más altos, mientras que las áreas azul oscuro y púrpura indican los valores más bajos.

Distribución de Precipitación:

Los puntos de observación (en rojo) están distribuidos en varias regiones, y la interpolación IDW calcula valores intermedios basados en estos puntos.
Se observa una concentración de puntos rojos en la parte central del mapa, lo que sugiere que hay una mayor densidad de datos de observación en esta región. Esto proporciona más precisión en la interpolación en esa zona en comparación con áreas más dispersas.

Patrones de Precipitación:

En la parte noreste (cerca de los puntos más densos en rojo), se observa un área de color amarillo que representa altos niveles de precipitación (cercanos a 105 mm). Esto sugiere que esta región puede tener condiciones de mayor precipitación.
En contraste, hay un área en el sureste (color púrpura oscuro) que muestra niveles de precipitación estimados más bajos, cercanos a -15 mm. Aunque valores negativos no suelen ser físicamente significativos, esto puede deberse a limitaciones en la interpolación en áreas de pocos datos.

Zonas Interpoladas vs. Zonas Observadas:

Las regiones sin puntos de observación están interpoladas en base a la distancia a los puntos cercanos. Esto significa que el modelo está haciendo suposiciones en esas áreas que pueden no reflejar valores exactos, especialmente en regiones alejadas de las observaciones, como en el sureste y el extremo norte del mapa.
Las áreas donde los puntos rojos están más dispersos pueden tener interpolaciones menos confiables. Por ejemplo, el suroeste y partes del noreste tienen menor cobertura de puntos.

Implicaciones para el Análisis de Precipitación:

La interpolación sugiere que hay una variabilidad espacial considerable en la precipitación estimada.
Las zonas de alta precipitación (en amarillo/verde) y baja precipitación (azul/púrpura) ayudan a identificar posibles regiones con condiciones climáticas diferenciadas, lo cual es útil para la gestión de recursos hídricos o la planificación agrícola.

```python
# Importar las librerías necesarias
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
from shapely.geometry import Point

# URL de la API
url = "https://www.datos.gov.co/resource/sbwg-7ju4.json"

# Cargar los datos desde la URL
data = pd.read_json(url)

# Tomar una muestra aleatoria de 200 registros
data_sample = data.sample(n=200, random_state=42)

# Crear geometría de puntos para geopandas
geometry = [Point(xy) for xy in zip(data['longitud'], data['latitud'])]
gdf = gpd.GeoDataFrame(data, geometry=geometry, crs="EPSG:4326")  # Establece CRS a WGS 84

# Reproyectar a CRS compatible con contextily para visualización de mapas base
gdf = gdf.to_crs(epsg=3857)

# Visualizar el mapa con los puntos
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, markersize=gdf['valorobservado']*5, color='blue', alpha=0.6, legend=True)

# Agregar mapa base de OpenStreetMap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Configuración de etiquetas
ax.set_title("Distribución Espacial de la Precipitación Observada en Colombia")
ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")
plt.show()
```

Los puntos azules en el mapa representan las ubicaciones geográficas donde se observaron valores específicos de precipitación (datos de valorobservado), según las coordenadas de latitud y longitud proporcionadas por los datos.

El tamaño de cada punto azul está escalado en función del valor de recepción observado.

Es decir:

Puntos más grandes : Representan lugares donde la precipitación observada fue mayor.
Puntos más pequeños : Representan lugares con menores niveles de precipitación.
Esto permite ver visualmente dónde hay más o menos precipitación en Colombia según los datos, ayudándonos a identificar áreas con patrones de lluvia más intensos o más débiles.
